<font size=4 color=green> 
    1 母函数、罗德里格斯公式
</font>

函数 $G(x,z)=e^{2xz - z^2}$ 在给定 $x$ 时，在整个 $z$ 平面是解析函数，以原点为参考点做泰勒展开：
\begin{equation}
G(x,z) = e^{2xz - z^2} = \sum\limits^\infty_{n=0} \frac{ H_n(x)}{ n!} z^n,
\label{HermiteGeneratingFunction}
\tag{1}
\end{equation}
其中，$H_n(x)$ 叫做厄米多项式。很快可以求出几个特殊情况，$n=0,1$ 时,
\begin{equation}
H_0(x) = G(x,z)|_{z=0} = 1,
\end{equation}
\begin{equation}
H_1(x) = \frac{ \partial G(x,z) }{ \partial z } |_{z=0} = 2x.
\end{equation}

通过母函数，还可以写出厄米多项式的罗德里格斯公式，
\begin{equation}
H_n(x) = \{ \frac{\partial^n G(x,z) }{ \partial z^n} \}_{z=0} 
     = \{ e^{x^2} \frac{ \partial^n e^{-(z-x)^2} }{ \partial (z-x)^n } \}_{z=0}
     = (-1)^n e^{x^2} \frac{ d^n e^{-x^2} }{ dx^n}.
\label{Hermite-Rodriguez}
\tag{2}
\end{equation}
从这个形式，也可以看出，$H_n(x)$是一个$n$阶多项式。

<font size=4, color=green> 
    2 级数形式
</font>

将母函数做级数展开，可以直接得到厄米多项式的级数形式，
\begin{equation}
e^{-z^2 + 2xz} = \sum\limits^\infty_{p=0} \frac{ (-z^2)^p }{ p!} \sum\limits^\infty_{q=0} \frac{ (2xz)^q }{ q! }
= \sum\limits^\infty_{n=0} z^n \sum\limits^{[n/2]}_{p=0} \frac{ (-1)^p }{p!} \frac{ (2x)^{n-2p} }{ (n-2p)! }.
\tag{3}
\end{equation}
与(1)式相对照，可以得到
\begin{equation}
H_n(x) = \sum\limits^{[n/2]}_{p=0} \frac{ (-1)^p n! }{p! (n-2p)!} (2x)^{n-2p}.
\tag{4}
\end{equation}
所以，$n$为奇数时，$H_n(x)$只有奇数阶级数，$H_n(x)$是奇函数；$n$为偶数时，$H_n(x)$只有偶数阶级数，$H_n(x)$是偶函数。

<font size=4, color=green> 
    3 递推公式
</font>

因为母函数 $G(x,z) = e^{x^2} e^{-(z-x)^2} = e^{2xz-z^2}$，可以通过计算偏导数，得到递推公式。

① $\frac{ \partial G(x,z) }{ \partial x} = 2z G$，将$G(x,z)$ 的级数展开式代入，有
\begin{equation}
\sum\limits^\infty_{n=0} \frac{ H'_n(x)}{n!} z^n = 2z \sum\limits^\infty_{n=0} \frac{ H_n(x) }{ n!} z^n,
\end{equation}
两边任意$n$阶系数必须相等，所以有
\begin{equation}
\frac{ H'_n(x) }{ n!} = \frac{ 2H_{n-1}(x)}{ (n-1)!},
\end{equation}
即
\begin{equation}
H'_n(x) = 2n H_{n-1}(x), n = 1,2,\cdots
\label{Hermite-recursion1}
\tag{5}
\end{equation}

② $\frac{ \partial G(x,z) }{ \partial x } = - (2z-2x) G$，将$G(x,z)$ 的级数展开式代入，有
\begin{equation}
\sum\limits^\infty_{n=0} \frac{ H_n(x) }{ (n-1)! } z^{n-1} = -2 \sum\limits^\infty_{n=0} \frac{ H_n(x) }{ n! } z^{n+1} + 2 \sum\limits^\infty_{n=0} \frac{ x H_n(x) }{ n! } z^n,
\end{equation}
即
\begin{equation}
\frac{ H_{n+1}(x) }{ n! } = -2 \frac{ H_{n-1} (x) }{ (n-1)! } + \frac{ 2x H_n(x) }{ n! },
\end{equation}
即
\begin{equation}
H_{n+1}(x) - 2x H_n(x) + 2n H_{n-1} (x) = 0.
\label{Hermite-recursion2}
\tag{6}
\end{equation}

<font size=4, color=green> 
    4 厄米微分方程
</font>

厄米多项式满足一个微分方程，可以由递推式(5,6)得出。
对公式(5)两边求导，得到
\begin{equation}
H''_n(x) = 2n H'_{n-1}(x) = 2n \cdot 2(n-1) H_{n-2}(x),
\tag{7}
\end{equation}
上式第二个等号又用了一次公式(5)。
对公式(6)做代换$n \rightarrow n-1$，得到
\begin{equation}
H_n(x) - 2xH_{n-1}(x) + 2(n-1)H_{n-2}(x) = 0,
\end{equation}
所以有 $2(n-1)H_{n-2}(x) = 2xH_{n-1}(x)-H_n(x)$，代入(7)得到
\begin{equation}
H''_n(x) = 4nxH_{n-1}(x) - 2nH_n(x) = 2xH'_n(x) - 2nH_n(x),
\end{equation}
第二个等号又使用了一次(5)。
这样，我们得到了厄米微分方程
\begin{equation}
H''_n(x) - 2x H'_n(x) + 2nH_n(x) = 0.
\tag{8}
\end{equation}

<font size=4, color=green> 
    5 正交性、模方
</font>

根据斯图姆-刘维尔理论，可以将厄米微分方程乘上一个因子，使它变成一个自伴方程，这个因子如下构造，
\begin{equation}
\frac{1}{p_0(x)} exp[ \int^x \frac{p_1(t)}{p_0(t)} dt ] = exp[ \int^x (-2t) dt ] = e^{-x^2},
\end{equation}
乘到厄米方程上，得到
\begin{equation}
e^{-x^2} H''_n(x) - 2x e^{-x^2} H'_n(x) + 2n e^{-x^2} H_n(x) = 0.
\end{equation}
将它看作自伴算子 $ \hat{L} = e^{-x^2} \frac{ d^2}{dx^2} - 2x e^{-x^2} \frac{d}{dx} $ 的本征值方程
\begin{equation}
\hat{L} H_n(x) + 2n e^{-x^2} H_n(x) = 0,
\end{equation}
本征值为$2n$，权重函数为$e^{-x^2}$。
对任意非负整数$m,n$，若$m \neq n $，有
\begin{equation}
e^{-x^2}( H_m(x) H'_n(x) - H_n(x) H'_m(x) )|^\infty_{-\infty} = 0,
\end{equation}
所以有正交性
\begin{equation}
\int^\infty_{-\infty} e^{-x^2} H_m(x) H_n(x) dx = 0.
\tag{9}
\end{equation}
而模方可以用罗德里格斯公式(2)得到，
\begin{equation}
\int^\infty_{-\infty} e^{-x^2} H_n(x) H_n(x) dx
= (-1)^n \int^\infty_{-\infty} \frac{d^n}{dx^n} e^{-x^2} H_n(x) dx
= \int^\infty_{-\infty} e^{-x^2} \frac{d^n}{dx^n} H_n(x) dx
= 2^n n! \sqrt{\pi}.
\tag{10}
\end{equation}
上式第一个等号使用了罗德里格斯公式(2)；第二个等号使用了$n$次分部积分，以及$e^{-x^2}|_\infty = e^{-x^2}|_{-\infty} = 0$；第三个等号使用了递推式(5)，以及$H_0(x)=1$，还有$\int^\infty_{-\infty} e^{-x^2} dx = \sqrt{\pi}$。

<font size=4, color=green> 
    6 厄米多项式的完备性(unchecked)
</font>

在一定条件下，厄米多项式$H_n(x)$在$(-\infty, \infty$上构成完备基矢，即满足相应条件的任意函数$f(x)$可以按厄米多项式展开
\begin{equation}
f(x) = \sum\limits^\infty_{n=0} c_n H_n(x).
\end{equation}

<font size=4, color=green> 
    7 厄米函数
</font>

上面，我们把$H_n(x)$叫做厄米多项式，并且推导了它的正交性和模方。利用厄米多项式的正交性和模方，也可以定义``<font color=blue>厄米函数</font>"，
\begin{equation}
\psi_n(x) = \frac{1}{ \sqrt{ 2^n n! \sqrt{\pi} } } e^{-x^2/2} H_n(x).
\end{equation}
从$H_n(x)$的正交性、模方，可以容易得到$\psi_n(x)$的正交归一性：
\begin{equation}
\int^\infty_{-\infty} \psi_n(x) \psi_m(x) dx = \delta_{nm}.
\end{equation}
利用$H_n(x)$的微分方程，容易证明$\psi_n(x)$满足微分方程
\begin{equation}
\psi''_n(x) + (2n+1-x^2) \psi_n(x) = 0.
\tag{11}
\end{equation}

<font size=4, color=green> 
    8 量子力学中的应用：一维谐振子
</font>

一维谐振子薛定谔方程为
\begin{equation}
- \frac{\hbar^2}{2m} \phi''(x) + \frac{1}{2} m \omega^2 x^2 \phi(x) = E \phi(x).
\end{equation}
即
\begin{equation}
\phi''(x) + (\frac{2mE}{\hbar^2} - \frac{ m^2 \omega^2 }{ \hbar^2 } x^2 ) \phi(x) = 0.
\end{equation}
记$\xi = \sqrt{\frac{m\omega}{\hbar}} x, \psi(\xi) = \phi(x) = \phi( \sqrt{ \frac{\hbar}{m\omega}} \xi )$，则有
\begin{equation}
\frac{d^2}{d\xi^2} \psi(\xi) + (\frac{2E}{\hbar \omega} - \xi^2) \psi(\xi) = 0.
\end{equation}
这个形式与(11)是完全一致的，所以，一维谐振子的本征值与本征函数为
\begin{equation}
E_n = (n+\frac{1}{2}) \hbar \omega, ~~ \psi_n(x)  = \frac{1}{ \sqrt{ 2^n n! \sqrt{\pi} } } e^{-x^2/2} H_n(x)..
\end{equation}
